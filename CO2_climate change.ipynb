To prepare the data for analysis, both files were first loaded into Pandas DataFrames using the `.read_csv()` function.

Since the data was originally in wide format‚Äîwith each year represented as a separate column‚Äîit was reshaped into long format using the `.melt()` function.  
This transformation allows for time series analysis, with each row now representing a single country-year observation.

Finally, the two reshaped DataFrames were merged using the `.merge()` function, combining forest area in km¬≤ and forest area as a percentage into a single, unified dataset.

````python
# loading dataset
df_km = pd.read_csv("forest_area_km.csv")
df_pct = pd.read_csv("forest_area_percent.csv")
````

````python
print(df_km.columns)
print(df_pct.columns)
````

````python
# Reshape km¬≤
df_km_long = pd.melt(df_km,
                     id_vars=["Country Name", "Country Code"],
                     var_name="Year",
                     value_name="Forest Area (km¬≤)")

# Reshape percent
df_pct_long = pd.melt(df_pct,
                      id_vars=["Country Name", "Country Code"],
                      var_name="Year",
                      value_name="Forest Area (%)")

# Merge both
df_combined = pd.merge(df_km_long,
                       df_pct_long,
                       on=["Country Name", "Country Code", "Year"])

# Check result
df_combined.head()
````

#### 4.3 Handling Countries with Excessive Missing Data

Firstly, to identify which countries had more than 50% missing data, we generated a summary table showing the number and percentage of missing values per country across both forest area columns.

Secondly, we used the `remove_high_missing_countries_from_df_copy()` function to clean the dataset. This function calculates the percentage of missing values for each country and removes any country where more than 50% of the data is missing. This step ensures the dataset remains statistically reliable and is not skewed by countries with limited or incomplete records.

We intentionally chose not to impute missing values. Filling missing entries with the mean would flatten country-specific patterns and ignore meaningful variation over time. Setting missing values to zero would falsely imply total deforestation, and backfilling would assume stability that may not be accurate. Keeping these values as `NaN` avoids introducing incorrect assumptions into the dataset.

We have removed data for years 1990 and 1991 given the persistent distortion of data

````python
# Count missing values per country
nulls_by_country = df_copy.groupby("Country Name")[["Forest Area (km¬≤)", "Forest Area (%)"]].apply(lambda x: x.isna().sum())

# Total missing
nulls_by_country["Total Missing"] = nulls_by_country.sum(axis=1)

# Total records per country
total_rows = df_copy.groupby("Country Name").size()
nulls_by_country["Total"] = total_rows * 2

# % missing
nulls_by_country["% Missing"] = (nulls_by_country["Total Missing"] / (total_rows *2)) * 100

# Sort by most missing
nulls_by_country_sorted = nulls_by_country.sort_values(by="Total Missing", ascending=False)

nulls_by_country_sorted.head(10)  # top 10
````

````python
def remove_high_missing_countries_from_df_copy(threshold=50):
    global df_copy

    # Count missing values per country
    nulls = df_copy.groupby("Country Name")[["Forest Area (km¬≤)", "Forest Area (%)"]].apply(lambda x: x.isna().sum())

    # Total missing
    nulls["Total Missing"] = nulls.sum(axis=1)

    # Total expected entries
    total = df_copy.groupby("Country Name").size()
    nulls["Total"] = total * 2

    # Percent missing
    nulls["% Missing"] = (nulls["Total Missing"] / nulls["Total"]) * 100

    # Filter countries
    valid_countries = nulls[nulls["% Missing"] <= threshold].index
    removed_countries = nulls[nulls["% Missing"] > threshold].index

    # Apply filter to df_copy
    df_copy = df_copy[df_copy["Country Name"].isin(valid_countries)].reset_index(drop=True)

  # Report
    removed_rows = total[removed_countries].sum()
    print(f"{removed_rows} rows were removed.")
    print(f"{len(removed_countries)} countries were excluded for having more than {threshold}% missing data.")
    print("\nRemoved countries:")
    for country in removed_countries:
        print(f"- {country}")


    return df_copy
````

````python
df_copy = remove_high_missing_countries_from_df_copy()
````

````python
# Drop problematic years
df_copy = df_copy[~df_copy["Year"].isin([1990, 1991])]
````

````python
# Get start and end year forest % per country
df_change = (
    df_copy.sort_values(['Country Name', 'Year'])
    .groupby('Country Name')['Forest Area (%)']
    .agg(['first', 'last'])
    .rename(columns={'first': 'Start %', 'last': 'End %'})
)

# Calculate % change
df_change['Change (%)'] = df_change['End %'] - df_change['Start %']

# Categorise the change
def categorize(change):
    if change <= -20:
        return 'High Loss'
    elif change <= -5:
        return 'Moderate Loss'
    elif change >= 20:
        return 'High Gain'
    elif change >= 5:
        return 'Moderate Gain'
    else:
        return 'Stable'

df_change['Category'] = df_change['Change (%)'].apply(categorize)

# Add region (merge from df_copy)
region_lookup = df_copy[['Country Name', 'Region']].drop_duplicates()
df_change = df_change.merge(region_lookup, on='Country Name', how='left')
````

Filter the DataFrame to examine the % forest area within a specified range (`filter_by_forest_percnt`). This approach allows us to analyse how different percentage areas and year vs actual area, is distributed based on their % forest area, providing insights into viewing options and trends

````python
def filter_by_forest_percent(df_copy, min_pct_forest_area, max_pct_forest_area):
    """
    Filter a DataFrame based on a specified range and return the first two entries.

    Parameters:
    dataframe (pandas.DataFrame): The DataFrame to sort and filter.
    min__pct_forest_area (int): The minimum Forest Area (%) threshold.
    max__pct_forest_area (int): The maximum Forest Area (%) threshold.

    Returns:
    pandas.DataFrame: The first five entries of the filtered DataFrame.
    """
    # Sort the DataFrame by 'forest_percent'
    sorted_df = df_copy.sort_values('Forest Area (%)')

    # Filter the DataFrame for entries with a forest_percent within the specified range
    filtered_selection = sorted_df[
        (sorted_df['Forest Area (%)'] > min_pct_forest_area) & (sorted_df['Forest Area (%)'] <= max_pct_forest_area)
    ]

    # Return the first five entries of the filtered selection
    return filtered_selection.head(100)
````

**Graph 5**: This line chart tracks the average annual change in forest area (%) across all countries from 1992 to 2021, and overlays a linear trendline to help interpret long-term patterns.
- The chart focuses on the direction and consistency of global deforestation by measuring how much forest area changes each year, on average.
- Including a trendline allows us to assess whether this rate of change is improving or worsening over time, and helps smooth out short-term noise.

By using annual % change rather than total area, this approach highlights momentum ‚Äî whether the global forest situation is accelerating, stabilizing, or beginning to recover. It provides a clearer sense of trajectory than snapshots from just two endpoints.

````python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# STEP 1: Load the forest area percentage dataset
df_pct = pd.read_csv('forest_area_percent.csv')

# STEP 2: Reshape from wide to long format
df_long = df_pct.melt(
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name='Forest_area_pct'
)

# STEP 3: Rename columns for consistency
df_long.rename(columns={'Country Name': 'Entity'}, inplace=True)
df_long['Year'] = df_long['Year'].astype(int)

# STEP 4: Calculate year-over-year change per country
df_long["Forest_area_pct_change"] = df_long.groupby("Entity")["Forest_area_pct"].diff()

# STEP 5: Calculate average change per year across all countries
avg_change = df_long.groupby("Year")["Forest_area_pct_change"].mean()

# STEP 6: Convert to NumPy arrays
years = avg_change.index.to_numpy().astype(int)
values = avg_change.values.astype(float)

# STEP 7: Fit linear regression trendline
slope, intercept, _, _, _ = linregress(years, values)
trendline = slope * years + intercept

# STEP 8: Plot
plt.figure(figsize=(10, 5))
plt.plot(years, values, marker="o", label="Avg Yearly Change")
plt.plot(years, trendline, color="red", linestyle="--", label=f"Trendline (slope: {slope:.4f})")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')

plt.title("Average Annual Change in Forest Area (%) with Trendline")
plt.ylabel("Average Yearly Change (%)")
plt.xlabel("Year")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````
