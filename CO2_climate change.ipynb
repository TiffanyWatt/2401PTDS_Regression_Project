#### 4.1 Handling Missing Data: 
To ensure the dataset was complete and ready for regression analysis, missing values were handled in two steps:  

1. **Median by Area (Country/Region):**  
   Missing values in numeric columns were first imputed using the **median within each `Area`**.  
   This preserves local patterns by ensuring each country or region is filled with values that best reflect its own distribution.

2. **Global Median Fallback:**  
If missing values remained after the group-level imputation (e.g., when a country had insufficient valid entries to calculate a median), these were filled with the global median of the corresponding feature. This guarantees no missing values remain in the dataset.

 

````python
# Fill missing values in numeric columns with the median per country
df_CO2_copy = df_CO2_copy.groupby("Area").apply(
    lambda group: group.fillna(group.median(numeric_only=True))
)

# Reset index (optional, depending on your grouping)
df_CO2_copy.reset_index(drop=True, inplace=True)

# Apply global median for any NaNs left
df_CO2_copy = df_CO2_copy.fillna(df_CO2_copy.median(numeric_only=True))
````

````python
# Missing values
df_CO2_copy.isnull().sum().sort_values(ascending=False).head(11)
````

#### 4.2 Converting Area to Numeric  

The `Area` column contains categorical values representing countries or regions.  
Since regression models require numeric input, this column was converted into numeric form using **one-hot encoding**. This allows the model to capture differences across areas without imposing an artificial ordering.  

We then also want to remove all spaces in the column headings as Libraries like `statsmodels` and `scikit-learn` often expect clean variable names without spaces or special characters.

````python
# Convert 'Area' to numeric using one-hot encoding
df_CO2_copy = pd.get_dummies(df_CO2_copy, columns=['Area'], drop_first=True)

# Check the updated dataframe
df_CO2_copy.head()
````

### 6.1b Fitting the model using `statsmodels.OLS`

#### Generating the regression string



````python
import statsmodels.api as sm

# Copy dataset
df_CO2_model = df_CO2_copy.copy()

# Dependent variable
y_name = 'Average_Temperature_¬∞C'

# Independent variables
X_names = [col for col in df_CO2_model.columns if col != y_name]

# Define X and y
X = df_CO2_model[X_names].astype(float).to_numpy()
y = df_CO2_model[y_name].astype(float).to_numpy()

# Add constant
X = sm.add_constant(X, has_constant='add')
````

````python
# Build the OLS formula string " y ~ X "
formula_str = y_name+" ~ "+" + ".join(X_names);
print('Formula:\n\t {}'.format(formula_str))
````

#### Fitting the model

````python
# Fit the OLS regression
model = sm.OLS(y, X)
fitted = model.fit()

# Output the summary
print(fitted.summary())
````

Findings:

- The model fits reasonably well (R¬≤ ~ 0.53) but multicollinearity is a concern.

- We will need to check correlations or calculate VIF (Variance Inflation Factor) to identify overlapping predictors.

- After removing or combining redundant variables, the model will likely become more stable and easier to interpret.

### Variable selection by correlation and significance

üîé With 264 predictor variables available, it is important to identify which ones contribute most to explaining variations in **Average_Temperature_¬∞C**.  

Two main approaches guide variable selection:  

1. **Correlation Analysis**  
   - Examine the **correlation coefficients** between `Average_Temperature_¬∞C` and each independent variable.  
   - Select those with the **strongest correlations** (both positive and negative), as they are more likely to influence temperature.  
   - This helps narrow down predictors before modeling.  

2. **Statistical Significance in Regression**  
   - From the regression output, consider predictors with **low p-values (typically < 0.05)**.  
   - These variables are statistically significant and have a measurable impact on temperature.  
   - Coefficients also reveal whether the relationship is positive or negative.  

‚û°Ô∏è By combining correlation strength with statistical significance, we can select a smaller, more reliable set of predictors. This reduces multicollinearity, improves model interpretability, and helps avoid overfitting.

````python
# Calculate correlations between predictor variables and the response variable
corrs = df_CO2_model.corr()['Average_Temperature_¬∞C'].sort_values(ascending=False)
````

````python
from scipy.stats import pearsonr

# Build a dictionary of correlation coefficients and p-values
dict_cp = {}

column_titles = [col for col in corrs.index if col!= 'Average_Temperature_¬∞C']

for col in column_titles:
    p_val = round(pearsonr(df_CO2_model[col], df_CO2_model['Average_Temperature_¬∞C'])[1],6)
    dict_cp[col] = {'Correlation_Coefficient':corrs[col],
                    'P_Value':p_val}
    
df_cp = pd.DataFrame(dict_cp).T
df_cp_sorted = df_cp.sort_values('P_Value')
df_cp_sorted[df_cp_sorted['P_Value']<0.05]
````

Now, we've got a **sorted list of the p-values and correlation coefficients** for each of the features, when considered on their own. 

Let's keep only the variables that have a significant correlation with the dependent variable. We'll put them into an independent variable DataFrame `X`:

````python
# The dependent variable remains the same:
y_data = df_CO2_model[y_name]  # y_name = 'Average_Temperature_¬∞C'

# Model building ‚Äì Independent Variable (IV) DataFrame
X_names = list(df_cp[df_cp['P_Value'] < 0.05].index)
X_data = df_CO2_model[X_names]
````

### Finding and removing the highly correlated features

We also need to look for predictor variable pairs which have a **high correlation with each other** to avoid **autocorrelation**.

````python
import numpy as np
import pandas as pd

target = "Average_Temperature_¬∞C"

# 1. Compute correlation matrix of predictors
X = df_CO2_model.drop(columns=[target])
corr_matrix = X.corr()

# 2. Find highly correlated pairs (off-diagonal)
threshold = 0.9
corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            var1 = corr_matrix.columns[i]
            var2 = corr_matrix.columns[j]
            corr_val = corr_matrix.iloc[i, j]
            corr_pairs.append((var1, var2, corr_val))

# 3. Create DataFrame of highly correlated pairs
high_corr_df = pd.DataFrame(corr_pairs, columns=["Variable_1", "Variable_2", "Correlation_between_vars"])

# 4. Add correlation of each variable with the target
corr_with_target = df_CO2_model.corr()[target]

high_corr_df["Var1_to_Target"] = high_corr_df["Variable_1"].map(corr_with_target)
high_corr_df["Var2_to_Target"] = high_corr_df["Variable_2"].map(corr_with_target)

# 5. Sort by absolute correlation strength
high_corr_df = high_corr_df.sort_values(by="Correlation_between_vars", key=lambda x: abs(x), ascending=False)

high_corr_df.head(20)
````

````python
# === Drop highly correlated redundant variables ==========================

to_drop_vars = [
    "Total_Population_-_Male",
    "Food_Household_Consumption",
    "Manure_Management",
    "Rural_population",
    "Crop_Residues",
    "Food_Packaging",
    "Fires_in_humid_tropical_forests",
    "Rice_Cultivation"
]

df_CO2_model = df_CO2_model.drop(columns=to_drop_vars, errors="ignore")

print("Dropped variables:", to_drop_vars)
print("Remaining shape:", df_CO2_model.shape)
````

## 5. Variable selection by variance thresholds

Variance thresholds remove **features whose values don't change much** from observation to observation. 

The objective here is to **remove** all features that have a **variance lower than** the selected **threshold**.

**Note:** Variance is dependent on scale, so the features will have to be **normalised** before implementing variance thresholding.

````python
# Separate data into independent (X) and independent (y) variables
X_names = list(df_CO2_copy.columns)
X_names.remove(y_name)
X_data = df_CO2_copy[X_names]
y_data = df_CO2_copy[y_name]
````

````python
# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_data)
X_normalize = pd.DataFrame(X_scaled, columns=X_data.columns)
````

### Variance threshold in Scikit Learn

To implement variance threshold in Scikit Learn we have to do the following:

- Import and create an instance of the `VarianceThreshold` class.
- Use the `.fit()` method to select a subset of features based on the threshold.

````python
from sklearn.feature_selection import VarianceThreshold

# Create VarianceThreshold object
selector = VarianceThreshold(threshold=0.005)

# Use the object to apply the threshold on data
selector.fit(X_normalize)
````

````python
# Get column variances
column_variances = selector.variances_

vars_dict = {}
vars_dict = [{"Variable_Name": c_name, "Variance": c_var}
             for c_name, c_var in zip(X_normalize.columns, column_variances)]
df_vars = pd.DataFrame(vars_dict)
df_vars.sort_values(by='Variance', ascending=False)
````

## 11. References <a class="anchor" id="chapter11"></a>

Maintaining clear and organized documentation is crucial for reproducibility and credibility. Harvard referencing, a widely recognized citation style, can be effectively applied to data science notebooks to ensure proper acknowledgment of sources and facilitate the tracing of data and methodologies back to their origins.

1. Citing Data Sources: When gathering datasets for analysis, it's essential to cite the origin of the data. This includes providing details such as the author, publication date, title, and source of the dataset. In your notebook, you can include inline citations for each dataset you use, following the Harvard referencing format. 

<div class="alert alert-block alert-warning">
<b>Example:</b> The World Bank. (2020). World Development Indicators. Retrieved from https://datacatalog.worldbank.org/dataset/world-development-indicators
</div>

2. Referencing Research Papers: Data science often involves building upon existing research. Whether you're utilizing algorithms, methodologies, or theories from academic papers, it's important to reference them properly. Include citations for relevant papers within your notebook, both in the narrative sections and within code comments where applicable.

<div class="alert alert-block alert-warning">
<b>Example:</b> Smith, J., & Doe, A. (2018). Predictive Modeling Techniques for Financial Forecasting. Journal of Data Science, 10(2), 123-135.
</div>

3. Code Libraries and Packages: Data science projects heavily rely on various libraries and packages for analysis and visualization. When utilizing third-party code, make sure to acknowledge the creators by referencing the relevant documentation or papers associated with those tools.

<div class="alert alert-block alert-warning">
<b>Example:</b> Matplotlib Development Team. (2020). Matplotlib: A 2D plotting library. Retrieved from https://matplotlib.org/
</div>

4. Acknowledging Scholarly Contributions: If you've consulted online forums, blogs, or other informal resources during your project, it's courteous to acknowledge these contributions. While Harvard referencing is traditionally used for academic sources, you can adapt it to reference online resources as well.

<div class="alert alert-block alert-warning">
<b>Example:</b> Kaggle. (2021). Titanic: Machine Learning from Disaster. Retrieved from https://www.kaggle.com/c/titanic
</div>
